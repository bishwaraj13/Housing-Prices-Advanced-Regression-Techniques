# Data Preprocessing

```python
# Feature 'MSSubClass' has numerical value, but it is a nominal data. So, converting it to categorical data
train['MSSubClass'] = train['MSSubClass'].apply(str)
test['MSSubClass'] = test['MSSubClass'].apply(str)
```

## Missing Values

```python
missing_total = all_data.isnull().sum().sort_values(ascending=False)
percent = (all_data.isnull().sum() * 100/all_data.isnull().count()).sort_values(ascending=False)
missing_data = pd.concat([missing_total, percent], axis=1, keys=['Total', 'Percent'])
missing_data[missing_data['Percent'] > 0]['Percent'].plot.bar(by = 'Percent', figsize=(10,4))
```

## Imputation
I have taken a different aproaches for imputing missing values in different features. Following are the approaches, along with the list of features that would oblige by the methodology of imputation:

### Replace the missing values with "None"
PoolQC, Alley, Fence, FireplaceQu, GarageType, GarageFinish, GarageQual, GarageCond, MasVnrType, BsmtQual, BsmtCond, BsmtExposure, BsmtFinType1, BsmtFinType2, MSSubClass, MiscFeatures

```python
for col in ('PoolQC', 'Alley', 'Fence', 'FireplaceQu', 'GarageType', 'GarageFinish',
            'GarageQual', 'GarageCond', 'MasVnrType', 'BsmtQual', 'BsmtCond', 
            'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2', 'MSSubClass', 'MiscFeature'):
    all_data[col] = all_data[col].fillna("None")
```

### Replace the missing values with 0
GarageArea, GarageCars, BsmtFinSF1, BsmtFinSF2, BsmtUnfSF, TotalBsmtSF, BsmtFullBath, BsmtHalfBath, MasVnrArea

```python
for col in ('GarageArea', 'GarageCars', 'BsmtFinSF1', 'BsmtFinSF2', 
            'BsmtUnfSF','TotalBsmtSF', 'BsmtFullBath', 'BsmtHalfBath', 'MasVnrArea'):
    all_data[col] = all_data[col].fillna(0)
```

### Replacing missing values of "GarageYrBlt" with values of "YearBuilt"

```python
all_data['GarageYrBlt'] = all_data['GarageYrBlt'].fillna(all_data['YearBuilt'])
```

### Replacing the missing values with median value of each neighbourhood
LotFrontage

```python
all_data["LotFrontage"] = all_data.groupby("Neighborhood")["LotFrontage"].transform(
    lambda x: x.fillna(x.median()))
```

### Replace the missing values with most frequently occuring value
MSZoning, Electrical, KitchenQual, Exterior1st, Exterior2nd, SaleType

```python
for col in ('MSZoning', 'Electrical', 'KitchenQual','Exterior1st', 'Exterior2nd', 'SaleType'):
    all_data[col] = all_data[col].fillna(all_data[col].mode()[0])
```

### Replace missing values of "Functional" with "Typ"

```python
all_data['Functional'] = all_data['Functional'].fillna('Typ')
```

### Deleting feature "Utilities"
### Remove this piece of code and put it somewhere else please

```python
all_data = all_data.drop(['Utilities'], axis = 1)
```

```python
train = all_data[:ntrain]
train['SalePrice'] = label
train['Id'] = train_id
test = all_data[ntrain:].reset_index().drop(['index'], axis = 1)
test['Id'] = test_id

quantitative = [f for f in train.columns if train.dtypes[f] != 'object']
quantitative.remove('SalePrice')
quantitative.remove('Id')
qualitative = [f for f in train.columns if train.dtypes[f] == 'object']
```

# Encoding

Encoding qualitative variables according to ordering based on mean of SalePrice.

```python
def encode(train_frame, all_data_frame, feature):
    ordering = pd.DataFrame()
    ordering['val'] = train_frame[feature].unique()
    ordering.index = ordering.val
    ordering['spmean'] = train_frame[[feature, 'SalePrice']].groupby(feature).mean()['SalePrice']
    ordering = ordering.sort_values('spmean')
    ordering['ordering'] = range(1, ordering.shape[0]+1)
    ordering = ordering['ordering'].to_dict()
    
    for cat, o in ordering.items():
        all_data_frame.loc[all_data_frame[feature] == cat, feature+'_E'] = o
    
qual_encoded = []
for q in qualitative:  
    encode(train, all_data, q)
    qual_encoded.append(q+'_E')
print(qual_encoded)
```

# Correlation

Now that we have a lot of ordinal variables, it is better to use Spearman Correlation. For the record, spearman correlation works better with variables, even when they do not have linear relation between them.

```python
def spearman(frame, features):
    spr = pd.DataFrame()
    spr['feature'] = features
    spr['spearman'] = [frame[f].corr(frame['SalePrice'], 'spearman') for f in features]
    spr = spr.sort_values('spearman')
    plt.figure(figsize=(6, 0.25*len(features)))
    sns.barplot(data=spr, y='feature', x='spearman', orient='h')
    
features = quantitative + qual_encoded
spearman(train, features)
```

### Define a cross-validation strategy
We use the cross_val_score function of Sklearn. However this function has not a shuffle attribut, we add then one line of code, in order to shuffle the dataset prior to cross-validation

```python
#Validation function
n_folds = 5

def rmsle_cv(model):
    kf = KFold(n_folds, shuffle=True, random_state=42).get_n_splits(train[quantitative + qual_encoded].values)
    rmse= np.sqrt(-cross_val_score(model, train[quantitative + qual_encoded].values, y_train, scoring="neg_mean_squared_error", cv = kf))
    return(rmse)
```

## Averaging Models

```python
class AveragingModels(BaseEstimator, RegressorMixin, TransformerMixin):
    def __init__(self, models):
        self.models = models
        
    # we define clones of the original models to fit the data in
    def fit(self, X, y):
        self.models_ = [clone(x) for x in self.models]
        
        # Train cloned base models
        for model in self.models_:
            model.fit(X, y)

        return self
    
    #Now we do the predictions for cloned models and average them
    def predict(self, X):
        predictions = np.column_stack([
            model.predict(X) for model in self.models_
        ])
        return np.mean(predictions, axis=1)   
```

```python
class StackingAveragedModels(BaseEstimator, RegressorMixin, TransformerMixin):
    def __init__(self, base_models, meta_model, n_folds=5):
        self.base_models = base_models
        self.meta_model = meta_model
        self.n_folds = n_folds
   
    # We again fit the data on clones of the original models
    def fit(self, X, y):
        kfold = KFold(n_splits=self.n_folds, shuffle=True, random_state=156)
        
        # Train cloned base models then create out-of-fold predictions
        # that are needed to train the cloned meta-model
        out_of_fold_predictions = np.zeros((X.shape[0], len(self.base_models) + 1))
        for i, model in enumerate(self.base_models):
            for train_index, holdout_index in kfold.split(X, y):
                model.fit(X[train_index], y[train_index])
                y_pred = model.predict(X[holdout_index])
                out_of_fold_predictions[holdout_index, i] = y_pred
                
        out_of_fold_predictions      
        # Now train the cloned  meta-model using the out-of-fold predictions as new feature
        print(out_of_fold_predictions.shape)
        self.meta_model.fit(out_of_fold_predictions, y)
        return self
   
    #Do the predictions of all base models on the test data and use the averaged predictions as 
    #meta-features for the final prediction which is done by the meta-model
    def predict(self, X):
        df = pd.DataFrame()
        # make predictions using all base models
        for i, model in enumerate(self.base_models):
            df[str(i)] = model.predict(X)
        print('df')
        print(df.shape)
        #meta_features = np.column_stack([
            #np.column_stack([model.predict(X) for model in base_models]).mean(axis=1)
            #for base_models in self.base_models_ ])
        #return self.meta_model_.predict(meta_features)
        return self.meta_model.predict(df)
```

```python
class StackingAveragedModels2(BaseEstimator, RegressorMixin, TransformerMixin):
    def __init__(self, base_models, meta_model, n_folds=4):
        self.base_models = base_models
        self.meta_model = meta_model
        self.n_folds = n_folds
   
    # We again fit the data on clones of the original models
    def fit(self, X, y):
        df = pd.DataFrame()
        
        kfold = KFold(n_splits=self.n_folds, shuffle=True, random_state=156)
        out_of_fold_predictions = np.zeros((X.shape[0], len(self.base_models)))
        for i, model in enumerate(self.base_models):
            for train_index, holdout_index in kfold.split(X, y):
                model.fit(X.iloc[train_index], y.iloc[train_index])
                y_pred = model.predict(X.iloc[holdout_index])
                out_of_fold_predictions[holdout_index, i] = y_pred
        # split data into train-test
        X_dev, X_test, y_dev, y_test = train_test_split(X, y)
        X_dev.sort_index(axis=1, inplace=True)
        X_test.sort_index(axis=1, inplace=True)
        # train all base models on train
        for i, model in enumerate(self.base_models):
            model.fit(X_dev, y_dev)
            df[str(i)] = model.predict(X_test)
        # train a blender model using the dataframe
        self.meta_model.fit(df, y_test)
        return self
   
    #Do the predictions of all base models on the test data and use the averaged predictions as 
    #meta-features for the final prediction which is done by the meta-model
    def predict(self, X):
        X.sort_index(axis=1, inplace=True)
        df = pd.DataFrame()
        # make predictions using all base models
        for i, model in enumerate(self.base_models):
            df[str(i)] = model.predict(X)
        # make predictions using blender model
        return self.meta_model.predict(df)
```
